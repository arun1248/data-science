```python
import matplotlib.pyplot as plt
import numpy as np
```

# plots 

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

```python
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o-.r')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
plt.scatter([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

Difference between **plot with o and scatter**

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', mec = 'r') # mec markeredgecolor
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', mfc = 'r') # mfc markerfacecolor
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r', mec = 'r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = '#00FF00', mec = '#0000FF')
plt.show()
```

- linestyle ls (dotted :, dashed --)
- color c (r, '#FF0000')
- linewidth lw '22.5'

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = '--')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls = ':')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
x = np.array([10, 20, 30, 40, 50])
y = np.array([100, 200, 300, 400, 500])

plt.plot(x, y)

plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")

plt.show()
```

```python
x = np.array([10, 20, 30, 40, 50])
y = np.array([100, 200, 300, 400, 500])

plt.plot(x, y)

plt.title("Race Data")
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")

plt.show()
```

```python
x = np.array([10, 20, 30, 40, 50])
y = np.array([100, 200, 300, 400, 500])

plt.plot(x, y)

plt.title("Race Data", loc = 'left')
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")

plt.show()
```
