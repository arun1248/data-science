```python
df = pd.read_csv('student-mat.csv', sep=';')
df
```

```python
plt.scatter(df.age, df.absences, alpha=0.3)
plt.xlabel('Age')
plt.ylabel('Absences')
```

```python
age_absences = df.groupby('age')['absences'].mean().reset_index()
age_absences
```

```python
age_absences = df.groupby('age')['absences'].mean().reset_index()

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(age_absences['age'], age_absences['absences'], color='skyblue')
plt.xlabel('Age')
plt.ylabel('Average Absences')
plt.title('Average Absences by Age')
plt.xticks(age_absences['age'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Style

```python
x = np.array([35, 15, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.style.use('dark_background')

plt.pie(x,autopct='%.2f%%')
plt.legend(mylabels)
plt.show()
```

```python
x = np.array([35, 15, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.style.use('Solarize_Light2')

plt.pie(x,autopct='%.2f%%')
plt.legend(mylabels)
plt.show()
```
